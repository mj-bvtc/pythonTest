```python
import pandas as pd

file = r"V:\Projects\35 East Wacker\TerraCotta\_Preliminary Models\2021\35 Wacker 2020 N + E_multisheet.xlsx"

df = pd.read_excel(file, sheet_name = "Main")


points = df[["x","y"]].values

def bounding_box(points):
    """get the min/max values of a point list"""
    x_coordinates, y_coordinates = zip(*points)

    return [(min(x_coordinates), 
             min(y_coordinates)), 
            (max(x_coordinates), 
             max(y_coordinates))]
centers = dict()

for group in df.group.unique():
    """Get a unique list of groups
    take the XY coordinates of all points in group
    draw the bounding box/rectangle around min/max points
    find the center per group
    add these centers to a dictionary
    pass the dictionary back into df"""
    points = df.loc[df.group == group, ["x","y"]].values
    box = bounding_box(points)
    x1,y1 = box[0]
    x2,y2 = box[1]
    center = [(x1 + x2)/2 , (y1 + y2)/2]
    centers[group] = center

def find_center(x):
    return centers[x]

df["group_center"] = df.group.apply(find_center)

df.loc[:, ["group", "group_center"]].sort_values("group")
```
