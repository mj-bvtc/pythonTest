```python
import pandas as pd

path = r"V:\Projects\35 East Wacker\Documents\011019 REMEDIATION MARKUPS\35 E Wacker West Elevation Survey\report\report 2019-02-18 1248.xlsx"
df = pd.read_excel(path)

df.head()
```

```python
import re
df["new"] = ""

df.new = df.layer.apply(lambda x: "NEW" in x)
df.head()
```

```python
df["co"] = ""

def find_co(x):
    pat = r"CO([0-9]+)"
    result = re.search(pat, x)
    if result:
        return result.group(1)

    

df.co = df.layer.apply(find_co)
df.head()
```

```python
df["priority"] = ""

def find_priority(x):
    pat = r"P([0-9]+)"
    result = re.search(pat, x)
    if result:
        return result.group(1)

    

df.priority = df.layer.apply(find_priority)
df.head()
```

```python
df["swing_drop"] = ""
def get_drop(x):
    
    pat = "drop\s([\d]+)"
    match = re.search(pat, x)

    if match:
        return match.group(1)

df.swing_drop = df.zones.apply(get_drop)

df.head()
```
