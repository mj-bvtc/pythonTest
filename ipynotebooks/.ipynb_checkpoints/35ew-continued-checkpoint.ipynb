```python
ctrac = pd.read_excel(ct)
rhino = pd.read_excel(rh)
st = pd.read_excel(st)

list(rhino)
```

```python
cols = ["priority_ctrac", "elevation", "swing_drop", "name", "new", "z", "guid", "co", "priority_survey"]
rhino[cols]
```

```python
rhino["mk_priority"] = ""

rhino.loc[rhino.swing_drop == "E1", "mk_priority"] = 1
rhino.loc[rhino.swing_drop == "S1", "mk_priority"] = 2
rhino.loc[rhino.swing_drop == "S5", "mk_priority"] = 3
rhino.loc[rhino.swing_drop == "W1", "mk_priority"] = 4
rhino.loc[rhino.swing_drop == "W2", "mk_priority"] = 5
rhino.loc[rhino.swing_drop == "W3", "mk_priority"] = 6
rhino.loc[rhino.swing_drop == "S4", "mk_priority"] = 7
rhino.loc[rhino.swing_drop == "S3", "mk_priority"] = 8
rhino.loc[rhino.swing_drop == "S2", "mk_priority"] = 9
```

```python
cols.append("mk_priority")

rhino = rhino.sort_values(["mk_priority","z"], ascending=[True, True])
rhino[cols]
```

```python
rhino["bv_survey"] = ""

def check_survey(x):
    if x not in ["S2", "S3", "S4"]:
        return True
    else:
        return False
    
rhino.bv_survey = rhino.swing_drop.apply(check_survey)

cols.append("bv_survey")
rhino[cols]
```

```python
rhino["co11_12"] = ""

def find_co11_12(x):
    if x in [11,12]:
        return True
    else:
        return False
    
rhino.co11_12 = rhino.co.apply(find_co11_12)

rhino[["co", "swing_drop", "guid", "bv_survey", "mk_priority", "co11_12"]].head()
```

```python
col = ["co", "swing_drop", "guid", "bv_survey", "mk_priority", "co11_12"]
col = ["bv_survey", "co11_12", "mk_priority","z", "swing_drop","co", "guid"] #better order

#sorting blocks on site by priority we should pack crates 

rhino = rhino.sort_values(["bv_survey", "co11_12", "mk_priority","z"], ascending=[False, True, True, True])

rhino[col].head()
```
