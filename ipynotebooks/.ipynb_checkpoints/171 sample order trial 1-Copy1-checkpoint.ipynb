```python
import pandas as pd

file = r"C:\Users\mkreidler\Desktop\PS171_expanded_groups.xlsx"

df = pd.read_excel(file)

df = df.sort_values(["y", "x"], ascending=[False,True])

#df.to_excel(r"C:\Users\mkreidler\Desktop\PS171_sampleorder.xlsx")

df["bumper"] = df.swing_drop +"-"+ df.floor.map(str) +"-"+ df.name

cols = ["bumper", "group" ]

#df["draw_order"] = df.reset_index

df.reset_index(inplace = True, drop=True )

df["order"] = df.index + 1

df
```

```python


def sort_drops(x):
    drop_sort = {"NE1":1,
             "N1":2,
             "N2":3,
             "N3":4,
             "N4":5,
             "N5":6,
             "N6":7,
             "N7":8,
             "NW1":9,
             "SW1":10,
             "S1":11,
             "S2":12,
             "S3":13,
             "S4":14,
             "S5":15,
             "S6":16,
             "S7":17,
             "S8":18,
             "S9":19,
             "SE1":20}
    
    return drop_sort[x]

df["drop_sort"] = df.swing_drop.apply(sort_drops)

df = df.sort_values(["floor", "drop_sort","y", "x", "group"], ascending=[False,True,False,True,True])
df.reset_index(inplace = True, drop=True )
df["new_order"] = df.index + 1
df
```
