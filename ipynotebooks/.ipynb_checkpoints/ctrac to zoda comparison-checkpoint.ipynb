```python
import pandas as pd


pl = r"C:\Users\mkreidler\Desktop\35 wacker ctrac to zoda\35w Packing List 2019-04-16.xlsx"
ct = r"C:\Users\mkreidler\Desktop\35 wacker ctrac to zoda\35w_units to look for in ctrac.xlsx" 

packing = pd.read_excel(pl)
ctrac = pd.read_excel(ct)

ctrac.head()

```

```python
packing.head()
```

```python
ctrac["cnt"] = ctrac.groupby("Block ID").cumcount() +1
ctrac["cc"] = ctrac["Block ID"] + "#" +ctrac.cnt.map(str)

ctrac.head()
```

```python
packing.head()
```

```python
packing["cnt"] = packing.groupby("Block Style").cumcount() + 1
packing["cc"] = packing["Block Style"] + "#" + packing.cnt.map(str)

packing.head()
```

```python
merge = ctrac.merge(packing, on="cc", how="outer")

not_in_crate = merge[merge["Block Style"].isna() == True]
not_in_crate = not_in_crate[not_in_crate["Block ID"].isna() == False]

not_in_crate.to_excel(r"C:\Users\mkreidler\Desktop\35 wacker ctrac to zoda\not_in_crate.xlsx")

print(len(not_in_crate))
not_in_crate[["Block ID", "Block Style"]].head()
```

```python
c = ctrac.groupby("Block ID").sum().reset_index()
packing["pack_qty"] = 1
p = packing.groupby("Block Style").sum().reset_index()

m = c.merge(p, left_on = "Block ID", right_on="Block Style", how="outer").fillna(0)

m = m[["Block ID", "UnitQuantity", "pack_qty"]]


m = m.rename(columns={"UnitQuantity": "ctrac_qty"})
m["delta"] = m.ctrac_qty - m.pack_qty

print(len(m))

m.head()
```
