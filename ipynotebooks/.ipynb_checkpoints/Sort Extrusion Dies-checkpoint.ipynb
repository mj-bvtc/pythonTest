```python
import shutil, errno
import os

import pandas as pd

def copyanything(src, dst):
    
    try:
        name = src.split("\\")[-1]
        shutil.copytree(src, dst)
        #print(f"{src} Copied  to {dst}")        
        print(f"{name} done")
    except OSError as exc: # python >2.5
        #print(f"Error copying file {src} to {dst}")
        if exc.errno == errno.ENOTDIR:
            shutil.copy(src, dst)
            
        elif exc.errno == errno.EEXIST:
            print(f"File exists continuing on:  {name}")
        else: raise
            
def move_tree(src, dst, delete=False):
    copyanything(src, dst)
    if delete == True:
        shutil.rmtree(src)
        #print(f"Deleted {src}")
            
def find_folder(pattern, root):
    pass

def organize_dies(pats):
    path = r"V:\Projects\EXTRUSION\Die Cleanup - MK"
    

    for pat in pats:

        dst = r"V:\Projects\EXTRUSION\Die Cleanup - MK\DELETED"
        results = []
        for dirpath, dirnames, filenames in os.walk(path):
            for dirname in dirnames:
                if pat in dirname:
                    src = os.path.join(dirpath, dirname)
                    #dst = os.path.join(dst, dirname)
                    #print(src)
                    results.append(src)
                    #move_tree(src, dst)
        for src in results:
            root = r"V:\Projects\EXTRUSION\Die Cleanup - MK\DELETED"
            dirname = src.split("\\")[-1]
            dst = os.path.join(root, dirname)
            #print(src)
            #print(dst)
            move_tree(src, dst, delete=True)
    
    
    
excel = r"V:\Projects\EXTRUSION\Die Cleanup - MK\Die_Organization_2018-09-05.xlsx"

df = pd.read_excel(excel)

data = df[df["task"] == "delete"]

blocks = data.block.tolist()
blocks
organize_dies(blocks)

print("process complete")

```
