```python
import pandas as pd

file = r"C:\Users\mkreidler\Desktop\FCCH_estimate.xlsx"
estimate = pd.read_excel(file)

estimate = estimate.fillna(0)
estimate = estimate.loc[estimate.Style != 0, ["Style","Qty of Units"]]
estimate = estimate.rename(index=str, columns={"Style":"name", "Qty of Units":"EstQty"})
estimate.head()
```

```python
file2 = r"C:\Users\mkreidler\Desktop\fcch_survey.xlsx"
survey = pd.read_excel(file2)
df = pd.read_excel(file2)
survey.reset_index(inplace=True)
survey.rename(index=str, columns={"index":"old_index"}, inplace=True)
survey
survey = survey.groupby("name").count().reset_index()
survey = survey.rename(index=str, columns={"old_index":"SurveyQty"})
survey.head()
```

```python
merge = estimate.merge(survey, on="name", how="outer")
merge = merge.fillna(0)
merge["delta"] = merge.SurveyQty - merge.EstQty
merge.head()
```
