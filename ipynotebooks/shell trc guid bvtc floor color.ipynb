```python
import pandas as pd

scope_path = r"V:\Projects\Shell House Building\Documents\Block Production Status\2019-01-23\scope.xlsx"
level_path = r"V:\Projects\Shell House Building\Documents\Block Production Status\2019-01-23\level.xlsx"


scope = pd.read_excel(scope_path)
level = pd.read_excel(level_path)

scope.head()
```

```python
duplicates = new_scope[new_scope.trc.duplicated()]
duplicates.head()
```

```python
no_dup = new_scope[~new_scope.trc.duplicated()]
no_dup.head()
```

```python
data = no_dup
```

```python
seta = data.level.isna()
seta.head()
```

```python
data[~seta].head()
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
data.loc[seta, "level"] = "A"
data.loc[data["level"] == "Ground", "level"] = "G"
data.loc[data["level"] == "Clock", "level"] = "C"

data.head()
```

```python
data["sort"] = ""
data.loc[data["level"] == 12, "sort"] = 1
data.loc[data["level"] == 11, "sort"] = 2
data.loc[data["level"] == 10, "sort"] = 3
data.loc[data["level"] == 9, "sort"] = 4
data.loc[data["level"] == 7, "sort"] = 5
data.loc[data["level"] == 6, "sort"] = 6
data.loc[data["level"] == 5, "sort"] = 7
data.loc[data["level"] == 4, "sort"] = 8
data.loc[data["level"] == 3, "sort"] = 9
data.loc[data["level"] == 2, "sort"] = 10
data.loc[data["level"] == 1, "sort"] = 11
data.loc[data["level"] == "G", "sort"] = 12
data.loc[data["level"] == "C", "sort"] = 13
data.loc[data["level"] == "A", "sort"] = 14
data = data.sort_values("sort")
```
