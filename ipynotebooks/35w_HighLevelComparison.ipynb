```python
import pandas as pd
import re

rh = r"C:\Users\mkreidler\Desktop\35w_HighLevel\rhino.xlsx"
ct = r"C:\Users\mkreidler\Desktop\35w_HighLevel\ctrac.xlsx"

### Both of these are filtered to be current work
### ctrac filter out: void, zextra, NIS, anything in priorities 6s_d4, P7-P13 (may miss some units in RE)
### rhino filter out: NO CO11,CO12, no faulty, no deduct, no s2, 3, 4 include RE include new
rhino = pd.read_excel(rh)
ctrac = pd.read_excel(ct)


rhino["cum_count"] = rhino.groupby("name").cumcount() + 1
rhino["cc"] = rhino.name + "#" + rhino.cum_count.map(str)



ctrac["cum_count"] = ctrac.groupby("Block ID").cumcount() + 1
ctrac["cc"] = ctrac["Block ID"] + "#" + ctrac.cum_count.map(str)

```

```python
merge_out = rhino.merge(ctrac, on = "cc", how="outer")

merge_rh = rhino.merge(ctrac, on = "cc", how="left")

merge_ct = rhino.merge(ctrac, on = "cc", how="right")

len(merge_out)
```
