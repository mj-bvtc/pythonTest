```python
import pandas as pd
import re

ct = r"C:\Users\mkreidler\Desktop\35w\ctrac.xlsx"
rh = r"C:\Users\mkreidler\Desktop\35w\rhino.xlsx"
st = r"C:\Users\mkreidler\Desktop\35w\site.xlsx"

ctrac = pd.read_excel(ct)
rhino = pd.read_excel(rh)
site = pd.read_excel(st)
```

```python
ctrac = ctrac[["Priority", "Block ID", "UnitQuantity"]]

#pd.Series.unique

ctrac.head()
```

```python
ctrac = ctrac.groupby(["Priority", "Block ID"]).sum().reset_index()
```

```python
ctrac.head()
```

```python
ctrac["elevation"] = ""
ctrac["re"] = ""

def find_re(x):
    if "RE" in x:
        return "RE"
    else:
        return
    
ctrac.re = ctrac.Priority.apply(find_re)

ctrac.head()
```

```python
ctrac["p"] = ""

def find_p(x):
    pat = r"([\d]*)[a-zA-Z]*"
    result = re.search(string = x, pattern=pat)
    if result:
        return result.group(1)
    
ctrac.p = ctrac.Priority.apply(find_p)

ctrac.head()
```

```python


def find_elev(x):
    pat = r"[\d]*([a-zA-Z]*)"
    result = re.search(string = x, pattern=pat)
    if result:
        return result.group(1)
    
ctrac.elevation = ctrac.Priority.apply(find_elev)
ctrac.head()
```

```python
ctrac.elevation.unique()
```

```python
ctrac.Priority.unique()
```

```python
ctrac["batch"] = ""

def find_batch(x):
    pat = r"[_]([\d]*)"
    result = re.search(string = x, pattern=pat)
    if result:
        return result.group(1)
    
ctrac.batch = ctrac.Priority.apply(find_batch)
```

```python
ctrac[ctrac.batch.isna() == False].head()
```

```python
ctrac["swing_drop"] = ""

def find_drop(x):
    pat = r"[_][D]([\d]*)"
    result = re.search(string = x, pattern=pat, flags=re.IGNORECASE)
    if result:
        return result.group(1)
    
ctrac.swing_drop = ctrac.Priority.apply(find_drop)
ctrac.swing_drop.unique()
```

```python
ctrac[ctrac.swing_drop.isna() == False].head()
```

```python
ctrac["re_qty"] = ""

ctrac.re_qty = ctrac[ctrac.re.isna() == False].UnitQuantity
ctrac[ctrac.re.isna() == False]
```

```python
ctrac.p = pd.to_numeric(ctrac.p)
ctrac.batch = pd.to_numeric(ctrac.batch)
ctrac.swing_drop = pd.to_numeric(ctrac.swing_drop)
```
