```python
import pandas as pd
import re

file = r"V:\Projects\35 East Wacker\Field Work\BVTC Survey MK RK 2019-03-04\Reconciliation\MAB 35 e wacker2.xlsx"

df = pd.read_excel(file)

df.head()
```

```python
df["co"] = ""

def find_co(x):
    pat = r"CO[\s]*([0-9]+)"
    result = re.search(pat, str(x))
    if result:
        return result.group(1)

    

df.co = df.layer.apply(find_co)
df.head()
```

```python
df["priority"] = ""

def find_priority(x):
    pat = r"P([0-9]+)"
    result = re.search(pat, str(x))
    if result:
        return result.group(1)

    

df.priority = df.layer.apply(find_priority)
df.head()
```

```python
import re
df["new"] = ""

df.new = df.layer.apply(lambda x: "NEW" in str(x))
df.head()
```

```python
df["faulty"] = ""
df["remediation"] = ""
df["deduct"] = ""



def find_faulty(x):
    pat = r"faulty"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return "x"

df.faulty = df.layer.apply(find_faulty)
#df[df.faulty.notnull()]

def find_remediation(x):
    pat = r"remediation"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return "x"

df.remediation = df.layer.apply(find_remediation) 


def find_deduct(x):
    pat = r"deduct"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return "x"

df.deduct = df.layer.apply(find_deduct) 


df.head()
```

```python
df["question"] = ""



def find_question(x):
    pat = r"\?"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return "x"
    
    
df.question = df.name.apply(find_question) 
df.head()
```

```python
df.co = pd.to_numeric(df.co)
df.priority = pd.to_numeric(df.priority)

#df.name = df.name.apply(str)    ###does not work
```

```python
def make_string(x):
    pat = r"\(([^)]*)\)"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)
    else:
        return "fail"
    
    
df.name = df.name.apply(make_string) 
df.head()
```

```python
file2 = r"V:\Projects\35 East Wacker\Field Work\BVTC Survey MK RK 2019-03-04\Reconciliation\MAB setting 2019-03-19.xlsx"

data = pd.read_excel(file2)

data = data[["zones","guid", "x","y","z"]]

data.head()
```

```python
data["floor"] = ""
data["elevation"] = ""
data["project"] = ""


def find_floor(x):
    pat = r"floor\s*(\d*)"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)

data.floor = data.zones.apply(find_floor)
#df[df.floor.notna()]

def find_elevation(x):
    pat = r"([a-zA-Z]*)\s*ELEVATION"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)

data.elevation = data.zones.apply(find_elevation)
#df[df.elevation.notna()]    


def find_project(x):
    pat = r"'project\s+(.*)'"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)
    else:
        print("not in project")
        
data.project = data.zones.apply(find_project)

data
```

```python
df = df.merge(data, how="left", on="guid")

df
```

```python
df["swing_drop"] = ""
def get_drop(x):

    pat = r"drop\s*([a-z]*\d*)"
    match = re.search(pat, str(x), flags=re.IGNORECASE)

    if match:
        #print("Found Match")
        return match.group(1)
    else:
        #print("Not Found")
        return None

df.swing_drop = df.zones.apply(get_drop)

df
```
