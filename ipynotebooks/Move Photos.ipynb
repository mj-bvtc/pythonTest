```python
import shutil, errno
import os

import pandas as pd


def find_folder(path, pat):

    results = []
    for dirpath, dirnames, filenames in os.walk(path):
        for dirname in dirnames:
            if pat in dirname:
                src = os.path.join(dirpath, dirname)
                results.append(src)
                
    return results
    

    
    
    
pics = r"V:\Projects\EXTRUSION\Die Cleanup - MK\_Die Photos\Rest of the photos"
path = r"V:\Projects\EXTRUSION\Die Cleanup - MK"

for dirpath, dirnames, filenames in os.walk(pics):
    for file in filenames:
        if file.endswith(".JPG"):
            die = os.path.splitext(file)[0]
            die_path = os.path.join(dirpath, file)
            folders = find_folder(path, die)
            
            if folders:
                for f in folders:
                    dst = os.path.join(f, die) + ".JPG"
                    shutil.copyfile(die_path, dst)
                    print(f"Copied file {die} to {dst}")
                os.remove(die_path)
                print(f"Deleted {die}")
                print()
                print()
                    
                    
        else:
            print("NOT A PICTURE!  ...moving on")

print("process complete")

```
