```python
import pandas as pd
import re

ct = r"C:\Users\mkreidler\Desktop\35 w apples to apples\ct19.xlsx"
rh = r"C:\Users\mkreidler\Desktop\35 w apples to apples\rh.xlsx"

ct = pd.read_excel(ct)
rh = pd.read_excel(rh)


rh[["name", "priority_ctrac", "elevation", "swing_drop"]].head()
```

```python
rh["pre"] = ""
rh["prior_digit"] = ""
rh["elev_letter"] = ""


def get_digit(x):
    x = str(x)
    if x == "SHIPPED":
        return "Shipped"
    else:
        pat = r"(\d{1,10})"
        result = re.search(pattern=pat, string=x)
        if result:
            return result.group(1)
        
        
def get_letter(x):
    x = str(x)
    return list(x)[0]

rh.prior_digit = rh.priority_ctrac.apply(get_digit)
rh.elev_letter = rh.swing_drop.apply(get_letter)

rh.pre = rh.prior_digit + rh.elev_letter
rh[["name", "priority_ctrac", "elevation", "swing_drop", "pre"]]

rh["cumcount"] = rh.groupby(["pre", "name"]).cumcount() +1


rh["concat"] = "(" + rh.pre + ")" + rh["name"] + "[" + rh["cumcount"].map(str) + "]"

rh.concat
rh = rh[["name", "priority_ctrac", "elevation", "swing_drop", "pre", "concat"]]

rh
```

```python
ct[["Block ID", "Priority", "UnitQuantity"]]

ct["pre"] = ""

def get_pre(x):
    if x == "SHIPPED":
        return "Shipped"
    else:
        pat = r"(\d{1,10}[a-zA-Z]+)"
        result = re.search(pattern=pat, string=x)
        if result:
            return result.group(1)
        
ct.pre = ct.Priority.apply(get_pre)
        
ct[["Block ID", "Priority", "UnitQuantity", "pre"]]

ct["val"] = 1

ct["cumcount"] = ct.groupby(["pre", "Block ID"]).cumcount() +1

ct[["Block ID", "Priority", "UnitQuantity", "pre", "cumcount"]]

ct["concat"] = "(" + ct.pre + ")" + ct["Block ID"] + "[" + ct["cumcount"].map(str) + "]"


ct = ct[["Block ID", "Priority", "UnitQuantity", "pre", "cumcount", "concat", "Void Status"]]

ct.head()
```

```python
merge = rh.merge(ct, on = "concat", how = "outer")

merge.to_excel(r"C:\Users\mkreidler\Desktop\35 w apples to apples\merge.xlsx")
```
