```python
import pandas as pd

file = r"C:\Users\mkreidler\Desktop\modified shell.xlsx"

ctrac = pd.read_excel(file, sheet_name="ctrac")
reg = pd.read_excel(file, sheet_name="register")

def clean(x):
    x = str(x).strip().upper()
    return x

def swap_e(x):
    result = str(x).replace("-E","-A")
    return result

reg.bv = reg.bv.apply(clean)
reg["bvtc"] = reg.bv.map(str) + "-" + reg.color
reg.bvtc = reg.bvtc.apply(swap_e)
reg = reg.sort_values("bvtc")
reg["instance"] = reg.groupby("bvtc").cumcount() + 1
reg["compare"] = reg.bvtc + "[" + reg.instance.map(str) + "]"


ctrac.bvtc = ctrac.bvtc.apply(clean)
ctrac.bvtc = ctrac.bvtc.apply(swap_e)
ctrac = ctrac.sort_values("bvtc")
ctrac["instance"] = ctrac.groupby("bvtc").cumcount() + 1
ctrac["compare"] = ctrac.bvtc + "[" + ctrac.instance.map(str) + "]"

```
