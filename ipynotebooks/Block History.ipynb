```python
import pandas as pd
import sys
import datetime

log_file = r"C:\Users\mkreidler\Desktop\PS 171 Block History.txt"
with open(log_file, "w") as f:
    
    print("Project PS 171", file=f)
    print(datetime.datetime.now(), file=f)
    print(file=f)

    file = r"C:\Users\mkreidler\Desktop\PS 171 Block Register and Tracker (2)  -- Block Report.xlsx"

    df = pd.read_excel(file)

    df = df.rename(index=str, columns={'Block Description': "description",
     'Estimating ID' : "est_id",
     'BVTC Block ID': "block_id",
     '# Units with Arch. Comments': "arch_qty",
     'Units Moved due to Smaller Assemblies and 5 units removed from scope': "sm_assembly_qty",
     '# Units with Adjusted Joints': "survey_qty",
     '# Units from Field Survey2':"existing_qty",
     '# Units from  Arch Drawings':"est_work_qty",
     '# Units From Estimate': "base_qty",
     'NOTES': "notes"})

    df = df[["description","est_id","block_id","base_qty",
             "est_work_qty","existing_qty","survey_qty",
             "sm_assembly_qty","arch_qty", "notes"]]

    for i, row in df.iterrows():
        est_id = row.est_id
        block_id = row.block_id

        base_qty = int(row.base_qty)
        est_work_qty = int(row.est_work_qty)
        existing_qty = int(row.existing_qty)
        survey_qty = int(row.survey_qty)
        sm_assembly_qty = int(row.sm_assembly_qty)
        arch_qty = int(row.arch_qty)



        quantities = [base_qty, est_work_qty, existing_qty,
                      survey_qty, sm_assembly_qty, arch_qty]

        labels = ["the base bid", 
                  "estimating markups, but missed on base bid", 
                  "existing terra cotta in the field",
                 "the survey", 
                  "small assembly modifications", 
                  "architect changes, pending change order"]

        origin_index = next((i for i, x in enumerate(quantities) if x), None)

        origin = labels[origin_index]

        if est_id == block_id:
            print(f"Block name {block_id} matches the estimate", file=f)

        else:
            if str(est_id) != "nan":
                print(f"Block name {block_id} originally named {est_id} on estimate", file=f)
            else:
                print(f"Block name {block_id} did not exist on the estimate", file=f)

        print(f"This block originates from {origin}", file=f)

        overall_delta = arch_qty - base_qty

        if overall_delta == 0:
            print(f"Latest quantity matches estimate", file=f)
        
        # original_qty
        print(f"Original quantity from base: {base_qty}", file=f)
        

        # find deltas
        d_estimate = est_work_qty - base_qty
        if d_estimate != 0:
            print(f"Added quantity of {d_estimate} from units missed on base bid", file=f)
        d_existing = existing_qty - est_work_qty
        if d_existing != 0:
            print(f"Added quantity of {d_existing} found existing on site", file=f)
        d_survey = survey_qty - existing_qty
        if d_survey != 0:
            print(f"Added quantity of {d_survey} post survey", file=f)  
        d_sm = sm_assembly_qty - survey_qty
        if d_sm != 0:
            print(f"Added quantity of {d_sm} found in small assemblies", file=f)
        d_arch = arch_qty - sm_assembly_qty
        if d_arch != 0:
            print(f"Added quantity of {d_arch} per architect", file=f)
        print(f"Latest quantity, pending change order: {arch_qty}", file=f)
        if str(row.notes) != "nan":
            print(f"Note: {row.notes}", file=f)
        print(file=f)
        
print("done")
```
