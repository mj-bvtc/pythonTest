```python
import pandas as pd
import re


file = r"C:\Users\mkreidler\Desktop\Test Setting Folder\Verizon Building\Multisheet_export.xlsx"
df = pd.read_excel(file, sheet_name="Main")

names = df.loc[:, ["name"]]
pat = r"([a-zA-Z]+)"
def get_alpha(x):
    result = re.search(pat, x)
    if result:
        return result.group(1)
pat2 = r"([0-9]+)"
def get_num(x):
    result = re.search(pat2, x)
    if result:
        return result.group(1)
    
names["prefix"] = names.name.apply(get_alpha)
names["instance"] = names.name.apply(get_num)

names
```
