```python
import re
df["new"] = ""

df.new = df.layer.apply(lambda x: "NEW" in str(x))
df.head()
```

```python
df["co"] = ""

def find_co(x):
    pat = r"CO[\s]*([0-9]+)"
    result = re.search(pat, str(x))
    if result:
        return result.group(1)

    

df.co = df.layer.apply(find_co)
df.head()
```

```python
df["priority_ctrac"] = ""

def find_priority(x):
    pat = r"P([0-9]+)"
    result = re.search(pat, str(x))
    if result:
        return result.group(1)

    

df.priority_ctrac = df.layer.apply(find_priority)
df.head()
```

```python
df["swing_drop"] = ""
def get_drop(x):

    pat = r"drop\s*([a-z]*\d*)"
    match = re.search(pat, str(x), flags=re.IGNORECASE)

    if match:
        #print("Found Match")
        return match.group(1)
    else:
        #print("Not Found")
        return None

df.swing_drop = df.zones.apply(get_drop)

df
```

```python
df["faulty"] = ""
df["remediation"] = ""
df["deduct"] = ""
df["floor"] = ""
df["elevation"] = ""
df["project"] = ""


def find_faulty(x):
    pat = r"faulty"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return "x"

df.faulty = df.layer.apply(find_faulty)
#df[df.faulty.notnull()]

def find_remediation(x):
    pat = r"remediation"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return "x"

df.remediation = df.layer.apply(find_remediation) 


def find_deduct(x):
    pat = r"deduct"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return "x"

df.deduct = df.layer.apply(find_deduct) 

def find_floor(x):
    pat = r"floor\s*(\d*)"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)

df.floor = df.zones.apply(find_floor)
#df[df.floor.notna()]

def find_elevation(x):
    pat = r"([a-zA-Z]*)\s*ELEVATION"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)

df.elevation = df.zones.apply(find_elevation)
#df[df.elevation.notna()]    


def find_project(x):
    pat = r"'project\s+(.*)'"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)
    else:
        print("not in project")
        
df.project = df.zones.apply(find_project)
df[df.project.notna()].head()      

```

```python

c = df[["guid", "z", "swing_drop"]]

class Order():
    def __init__(self):
        self.swing_drop = None
        self.priority_survey = None
        self.sort = None

def create_order():
    drops = ["W3", "S1", "E1", "W2", "S5", "W1", "S2", "S3","S4"]
    priority_survey = [1, 1, 1, 2, 2, 2, 3, 3, 3]
    order = []
    for i in range(9):
        j = i + 1
        k = Order()
        k.swing_drop = drops[i]
        k.priority_survey = priority_survey[i]
        k.sort = j
        order.append(k)
        
    return order
        

o = create_order()

order = pd.DataFrame([vars(f) for f in o])
order
c.head()
```

```python
c = c.merge(order, on="swing_drop")
c.head()
```
