```python
import pandas as pd
import numpy as np
import re


rh = r"C:\Users\mkreidler\Desktop\new100Hudson_Zone_output.xlsx"
rhino = pd.read_excel(rh)

rhino.head()
```

```python
rhino["swing_drop"] = ""
rhino["floor"] = ""
rhino["elevation"] = ""
rhino["project"] = ""
rhino["ct_val"] = 1
rhino["ct_order"] = ""


def get_drop(x):
    pat = r"drop\s*([a-z]*\d*)"
    match = re.search(pat, str(x), flags=re.IGNORECASE)

    if match:
        # print("Found Match")
        return match.group(1)
    else:
        # print("Not Found")
        return None
    

def find_floor(x):
    pat = r"floor\s*(\d*)"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return int(result.group(1))    
    
def find_elevation(x):
    pat = r"ELEVATION\s*([a-zA-Z\s]*)"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)


def find_project(x):
    pat = r"'project\s*([a-zA-Z\s\d\(\)]*)'"
    result = re.search(pat, str(x), flags=re.IGNORECASE)
    if result:
        return result.group(1)
    else:
        print("not in project")
        

rhino.swing_drop = rhino.zones.apply(get_drop)
rhino.floor = rhino.zones.apply(find_floor)
rhino.elevation = rhino.zones.apply(find_elevation)
rhino.project = rhino.zones.apply(find_project)

rhino.floor = pd.to_numeric(rhino.floor)


rhino = rhino.sort_values(["floor", "swing_drop", "name" ], ascending=[True,True,True])
rhino.ct_order = rhino.ct_val.cumsum()

rhino.to_excel(r"C:\Users\mkreidler\Desktop\new100Hudson_Expanded.xlsx")
```
