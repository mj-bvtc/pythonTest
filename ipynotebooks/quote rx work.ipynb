```python
import pandas as pd
import numpy as np

file = r"C:\Users\mkreidler\Desktop\Test Setting Folder\Multisheet_export.xlsx"

df = pd.read_excel(file, sheet_name="Main")

df
```

```python
data = df[["name", "guid"]].groupby("name").count().sort_values("guid", ascending=False).reset_index()

data.head()
```

```python
import math

def assign_method(x):
    if x > 50:
        return "EXT"
    elif 50>=x>=30:
        return "RAM"
    else:
        return "HP"
    
    
data["form_method"] = data.guid.apply(assign_method)
        
#data["Dies/Molds"] = None

#data.loc[data.form_method == "EXT", "Dies/Molds"] = 1

#df['col_3'] = df.apply(lambda x: f(x.col_1, x.col_2), axis=1)

def get_dies(form, qty):
    if form == "EXT":
        return math.ceil(qty/50)
    elif form == "RAM":
        return math.ceil(qty/30)
    else:
        return math.ceil(qty/12)

data["Dies/Molds"] = data.apply(lambda x: get_dies(x.form_method, x.guid), axis=1)

data
```
