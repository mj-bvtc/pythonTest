```python
import pandas as pd


file = r"C:\Users\mkreidler\Desktop\PACKING LIST ZODA 353 W.xlsx"

df = pd.read_excel(file)

def delete_underscore(x):
    if str(x) not in ["","nan"]:
        return str(x).replace("_", "")
    return None

df["Block Style"] = df["Block Style"].apply(delete_underscore)

df.to_csv(r"C:\Users\mkreidler\Desktop\convert.txt")
```
