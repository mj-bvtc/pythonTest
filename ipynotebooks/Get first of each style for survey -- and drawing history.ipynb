```python
import datetime
import workdays
import getpass
import socket
import uuid
import random
import time

def strTimeProp(start, end, format, prop):
    """Get a time at a proportion of a range of two formatted times.

    start and end should be strings specifying times formated in the
    given format (strftime-style), giving an interval [start, end].
    prop specifies how a proportion of the interval to be taken after
    start.  The returned time will be in the specified format.
    """

    stime = time.mktime(time.strptime(start, format))
    etime = time.mktime(time.strptime(end, format))

    ptime = stime + prop * (etime - stime)

    return time.strftime(format, time.localtime(ptime))


def randomDate(start, end, prop):
    return strTimeProp(start, end, '%m/%d/%Y %I:%M %p', prop)



class Common():
    def __init__(self):
        self.user = getpass.getuser()
        self.device = socket.gethostname()
        self.created_dt = datetime.datetime.now()
        self.guid = uuid.uuid4()
    
    def report(self):
        message = f"Object created at {self.created_dt} by user {self.user} on {self.device} machine with guid {self.guid}"
        return message
    

class Drawing(Common):
    def __init__(self):
        super().__init__()
        self.modeled_dt = None
        self.first_drawn_dt = None
        self.submittals_dt = []
        self.approvals_dt = []
        self.modeler = None
        self.drafter = None
        self.info_available_dt = None
        self.sample = None
        self.notes = None
        self.name = None
        self.project_name = None
        self.project_number = None
        
        
drawings = []
examples = ["BK7", "CS12", "H9R"]

for e in examples:
    d = Drawing()
    d.name = e
    for i in range(3):
        d.submittals_dt.append(randomDate("1/1/2008 1:30 PM", "12/31/2019 4:50 AM", random.random()))
    drawings.append(d)

dd = pd.DataFrame([vars(f) for f in drawings])

dd.to_excel(r"C:\Users\mkreidler\Desktop\test_drawing.xlsx")
```
