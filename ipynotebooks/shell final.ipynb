```python
import pandas as pd

file = r"C:\Users\mkreidler\Desktop\shell final.xlsx"

mc = pd.read_excel(file, sheet_name="register")
ct = pd.read_excel(file, sheet_name="ctrac")

def clean(x):
    a = str(x).upper()
    b = a.strip()
    c = b.replace("-E", "-A")
    return c

mc.block = mc.block.apply(clean)
ct.block = ct.block.apply(clean)
mc["mc"] = 'mc'
ct["ct"] = 'ct'



merge = mc.merge(ct, on="block", how="outer")

find = merge[(merge.mc.isna() == False)&(merge.ct.isna())]

find
```
