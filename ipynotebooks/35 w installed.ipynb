```python
import pandas as pd

sh = r"V:\Projects\35 East Wacker\Field Work\BVTC Survey MK RK 2019-03-04\Spreadsheets\Installed\shipped_ctrac.xlsx"
inst = r"V:\Projects\35 East Wacker\Field Work\BVTC Survey MK RK 2019-03-04\Spreadsheets\Installed\installed_ctrac.xlsx"


ship = pd.read_excel(sh)
install = pd.read_excel(inst)

ship = ship.groupby("name").count().add_suffix("_ship-sum").reset_index()
```
