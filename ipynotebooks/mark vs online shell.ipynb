```python
import pandas as pd


mc = r"C:\Users\mkreidler\Desktop\mark.xlsx"
onl = r"C:\Users\mkreidler\Desktop\online.xlsx"
ctrac = r"C:\Users\mkreidler\Desktop\ctrac.xlsx"

mark = pd.read_excel(mc)
online = pd.read_excel(onl)
mark["bvtc"] = mark.bv + "-" + mark.color
online["bvtc"] = online.bv + "-" + online.color
mark["instance"] = mark.groupby("bvtc").cumcount()+1
online["instance"] = online.groupby("bvtc").cumcount()+1
mark["compare"] = mark.bvtc + "[" + mark.instance.map(str) + "]"
online["compare"] = online.bvtc + "[" + online.instance.map(str) + "]"
mark.head()
```

```python
merge = mark.merge(online, on="compare", how="outer")

merge.head()
```
