```python
import pandas as pd
from reportlab.pdfgen import canvas
import qrcode
import uuid
import datetime

### GET DATA
stuff = list("abcdefghijklmnopqrstuvwxyz")
df = pd.DataFrame({'stuff':stuff})
df["num"] = df.index + 1
df["c"] = df.num % 3
df.c = df.c.replace(0,3)
df["page"] = ((df.num + 23) // 24)
df["r"] = (((df.num-1)//3)+1)%8
df.r = df.r.replace(0,8)

### FILE LOCATIONS
qr_path = r"C:\Users\mkreidler\Desktop\pdf print barcodes\test.png"
logo_path = r"C:\Users\mkreidler\Desktop\pdf print barcodes\BVTC-Logo-BLACK-horizontal.png"
pdf_suffix = r"C:\Users\mkreidler\Desktop\pdf print barcodes\\"
dt = datetime.datetime.now()
ext = ".pdf"
pdf_path = f"{pdf_suffix}{dt}{ext}".replace(":","_")

### MAKE CANVAS
c = canvas.Canvas(pdf_path)
c.setPageSize((612,792))

### SPLIT INTO PAGES, COLUMNS, ROWS
for index, row in df.iterrows():
    
    page = row["page"]
    col = row["c"]
    _row = row["r"]
    i = row["stuff"]
    code = f"{i}"
    print(f"page {page} row {_row} column {col} info {i}")





    ### DRAW 1 TAG
    
    def draw_tag(x_num, y_num, code, qr_path):
        
        x_increment = 204
        y_increment = 99

        x = x_increment * (x_num-1)
        y = y_increment * (y_num-1)
        
        c.drawString(102.5491 + x, 33.7174+ y, code)
        c.drawInlineImage(qr_path, 0+ x,0+ y, width=99, height=99) 
        c.drawInlineImage(logo_path, 102.5491+ x,71.2396+ y, width=84.2926, height=14.0488) 



### COPY ACROSS PAGE 1, ROWS, COLUMNS

### ADD PAGE
c.showPage()

### COPY ETC...

### SAVE DOCUMENT

c.save()
```
