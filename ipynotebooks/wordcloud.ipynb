```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator


file = r"V:\Projects\35 East Wacker\Field Work\BVTC Survey #2\Multisheet_export.xlsx"

df = pd.read_excel(file, sheet_name="Main")

string_value_list = df.name.values
string_value_list = [x.strip() for x in string_value_list]
text = " ".join(string_value_list)

wordcloud = WordCloud(width=1000, height=500, background_color="white", collocations=False).generate(text)

# Display the generated image:
plt.figure(figsize=[36,24])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.show()

save = r"C:\Users\mkreidler\Desktop\Test Setting Folder\word_cloud.png"

wordcloud.to_file(save)
```
