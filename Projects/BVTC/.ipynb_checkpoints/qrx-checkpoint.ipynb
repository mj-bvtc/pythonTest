```python
def get_molds(presses, stock):
    """
    Iterates through mold purchase combinations
    to notify you if you need to purchase molds
    or if previous stock will suffice
    """
    press_per_mold = 12  #constant 12 presses per mold, determined with BVTC experience
    buy = 0
    if presses > stock:                           # check if you have molds in stock/inventory
        deficit = presses - stock                 # see how many presses you need to cover between stock and # units
        buy = (deficit - 1) // press_per_mold + 1
        stock += buy*press_per_mold                # increase stock by purchase * 12
        stock -= presses                          # remove pressed units from stock
    else:
        stock -= presses                          #remove presses from stock/inventory
    return buy, stock
```
